```python
import csv
import numpy as np
import matplotlib.pyplot as plt
```

```python
with open('test_results/varying_N_and_M.csv', newline='') as csvfile:
    data = list(csv.reader(csvfile))
for i in range(len(data)):
    data[i] = np.asarray(data[i][1:], dtype=float)

entries = len(data[0])

# varying N IDG
plt.rcParams.update({'font.size': 25})
fig = plt.figure(figsize = (8, 4))

z = np.polyfit(data[0], data[1], 6)
p = np.poly1d(z)
plt.plot(data[0], p(data[0]))
plt.ylabel("Deletion Score")
plt.xlabel("N")
plt.ylim([0.0, 0.2])
plt.show()

# modifying M IDG
plt.rcParams.update({'font.size': 25})
fig = plt.figure(figsize = (8, 4))

z = np.polyfit(data[0], data[2], 5)
p = np.poly1d(z)
plt.plot(data[0], p(data[0]))
plt.ylabel("Deletion Score")
plt.xlabel(" M")
plt.ylim([0.0, 0.2])
plt.show()
```

```python
# IDG approx error
with open('test_results/idg_approx_60_divs.csv', newline='') as csvfile:
    data = list(csv.reader(csvfile))

length = len(data[0]) 

for i in range(1, length):
    data[0][i] = int(data[0][i])
    data[1][i] = float(data[1][i])
    data[2][i] = float(data[2][i])

steps = np.array(data[0][1:length])
integral_error = np.array(data[2][1:length])
gnd_truth = float(data[3][1])

### attribution error ####
plt.rcParams.update({'font.size': 20})
fig = plt.figure(figsize = (8, 4))
plt.plot(steps, integral_error)
plt.axhline(0, color = "black")
plt.xlabel("Number of Steps")
plt.ylabel("Attribution Error")
plt.title("Attribution Error vs. Steps Used")
plt.show()
```
